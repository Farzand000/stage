```python
import pandas as pd

files = ["b2b_cleaned.csv","b2b_poland_1.csv"]


clean_final_dfs = []

for f in files:
    final_df = pd.read_csv(
        f,
        encoding='latin-1',          # tolerant encoding
        encoding_errors='replace'    # prevents crash
    )
    clean_final_dfs.append(final_df)

final_final_df = pd.concat(clean_final_dfs, ignore_index=True)

final_df["Created at"] = pd.to_datetime(final_df["Created at"], errors="coerce", utc=True)
# Alleen timezone verwijderen als het echt timezone-aware is
if final_df["Created at"].dt.tz is not None:
    final_df["Created at"] = final_df["Created at"].dt.tz_localize(None)

final_df["Jaar"] = final_df["Created at"].dt.year
final_df["Maand"] = final_df["Created at"].dt.month
final_df["Dag"] = final_df["Created at"].dt.day

```

```python
# Verwijder rijen waar Shipping Protection voorkomt
final_df = final_df[~final_df["Lineitem name"].str.contains(
    r"Patta Shipping Protection\s-\s\d+",
    regex=True,
    na=False)]

final_df
```

```python
# Save clean UTF-8 version
final_df.to_csv('wholesale_ss26.csv', index=False, encoding='utf-8')

print("Saved successfully.")
```

```python
import pandas as pd

# Inladen
df = pd.read_csv("global_mainline.csv")

# Zorg dat postcode string is
df["Shipping Zip"] = df["Shipping Zip"].astype(str)

# Selecteer Nederlandse records
mask_nl = df["Shipping Country"].isin(["NL", "Netherlands", "Nederland"])

# NL postcodes omzetten naar PC4
df.loc[mask_nl, "Shipping Zip"] = (
    df.loc[mask_nl, "Shipping Zip"]
        .str.replace(" ", "", regex=False)
        .str.upper()
        .str[:4]
)

# Opslaan als nieuwe CSV
df.to_csv("global_mainline_pc4.csv", index=False)

print("Nieuwe CSV aangemaakt: global_mainline_pc4.csv")
```
