```python
import pandas as pd

df_uk = pd.read_csv("uk.csv")

# Verwijder rijen waar Shipping Protection voorkomt
df_uk = df_uk[~df_uk["Lineitem name"].str.contains(
    r"Patta Shipping Protection\s-\s\d+",
    regex=True,
    na=False
)]

df_uk.to_csv("uk_new.csv", index=False)

print("Shipping Protection rijen verwijderd.")
```
